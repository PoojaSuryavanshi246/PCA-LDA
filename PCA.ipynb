#### A. Feature Elimination:
Removes columns not rows.
Removes irrelevant or redundant features ‚Äî keeps original features, just fewer.
1. remove multicollinearity
2. computational performance improves

#### B. Feature Extraction
Creates new features by transforming original ones ‚Äî loses interpretability.

### **Principal Component Analysis (PCA)**

eg: 2D to 1D


### ‚úÖ **PCA (Principal Component Analysis)**

- **Unsupervised** technique ‚Äî no class labels needed  
- Goal: **Maximize variance** in the data ‚Äî find directions (PCs) that capture most spread  
- Finds **orthogonal principal components** (PC1, PC2, ‚Ä¶) ‚Äî each explains decreasing variance  
- Output dimensions can be as many as original features (but usually reduced)  
- Best for:  
  - Data compression  
  - Noise reduction  
  - Visualization (2D/3D plots of high-D data)  
  - Preprocessing for clustering or regression  
- Assumes:  
  - Linear relationships between features  
  - Features are scaled (standardized) for fair comparison  
- Doesn‚Äôt care about classes ‚Äî only structure and spread  
- Common use:  
  - Face recognition (Eigenfaces)  
  - Genomics (gene expression patterns)  
  - Image/signal compression  
  - Financial risk modeling (reducing asset factors)  
  - Customer segmentation (behavioral features)  
- Often used *before* unsupervised learning or as a preprocessing step  
- Can be combined with LDA (e.g., PCA first to reduce noise, then LDA for classification)  
- **Does NOT guarantee class separation** ‚Äî may even mix classes if they share variance  

---

üìå **Remember:**  
**PCA = ‚ÄúShow me the shape of my data‚Äù**  
**LDA = ‚ÄúShow me how to tell groups apart‚Äù**



#### **How do you do a Principal Component Analysis**

**Step1**: Standardize the range of continuous initial variables.

**Step2**: Compute the covariance matrix to identify correlation.

**Step3**: Compute the eigenvectors and eigenvalues of the covariance matrix to identify principal components.

**Step4**: Sort the eigen values in descending order to pick the highest one.

**Step5**: Recast the data along the principal component axes.

### Summarizing the PCA end to end

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. Load data
digits = load_digits()
X, y = digits.data, digits.target

# 2. Split train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20, stratify=y)

# 3. Standardize features (CRITICAL for PCA)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # Use same scaler on test

# 4. Apply PCA to retain 95% of variance
pca = PCA(0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# 5. Train classifier on reduced features
clf = LogisticRegression(max_iter=1000, random_state=42)
clf.fit(X_train_pca, y_train)

# 6. Predict and evaluate
y_pred = clf.predict(X_test_pca)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Components retained: {pca.n_components_}")
```

### Note:
Scale after split to prevent test data from influencing training statistics ‚Äî avoiding data leakage and ensuring realistic model evaluation.

### High level points
---


1. Load  `X, y = load_digits().data, load_digits().target`
2. Split train/test: `train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)`
3. Scale training data only: `scaler.fit_transform(X_train)`
4. Apply same scaler to test data: `scaler.transform(X_test)` *(never fit on test!)*
5. Apply PCA to train: `pca.fit_transform(X_train_scaled)` ‚Üí retains 95% variance
6. Transform test data with same PCA: `pca.transform(X_test_scaled)`
7. Train classifier (e.g., LogisticRegression) on `X_train_pca` and `y_train`
8. Predict on `X_test_pca` ‚Üí `y_pred = clf.predict(X_test_pca)`
9. Evaluate: `accuracy_score(y_test, y_pred)`  

---

### üö´ **Never Do This**
- Don‚Äôt scale or fit PCA on the full dataset before splitting  
- Don‚Äôt fit scaler or PCA on test data  
- Don‚Äôt use `fit_transform()` on test set ‚Äî always use `transform()`

---

### üí° Key Rule:
> **Fit on train. Transform train and test.**  
> That‚Äôs how you avoid leakage. That‚Äôs how you get real accuracy.
